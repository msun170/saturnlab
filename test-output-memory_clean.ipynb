```python
# Test 3: Large DataFrame table (>50 rows, should be truncated)
import pandas as pd
import numpy as np

df = pd.DataFrame(
    np.random.randn(200, 6),
    columns=['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon', 'Zeta']
)
df
```

```python
# Test 4: Small DataFrame (should NOT be truncated)
small_df = pd.DataFrame({'x': range(10), 'y': range(10)})
small_df
```

```python
# Test 5: Multiple matplotlib plots (test lazy image loading)
# Scroll down slowly - images should load as they enter viewport
import matplotlib.pyplot as plt

for i in range(5):
    fig, ax = plt.subplots(figsize=(6, 3))
    x = np.linspace(0, 10, 100)
    ax.plot(x, np.sin(x + i), label=f'sin(x + {i})')
    ax.set_title(f'Plot {i+1} of 5')
    ax.legend()
    plt.show()
```
