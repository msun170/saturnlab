```python
# Test 2: Pandas DataFrame (deep sizing)
import pandas as pd
import numpy as np

df_small = pd.DataFrame({'a': range(100), 'b': range(100)})
df_big = pd.DataFrame(np.random.randn(10000, 50))
print(f'df_small: {df_small.shape}')
print(f'df_big: {df_big.shape}')
```

```python
# Test 3: NumPy arrays (nbytes)
small_arr = np.zeros(100)
big_arr = np.random.randn(1000, 1000)
huge_arr = np.random.randn(5000, 5000)
print(f'small_arr: {small_arr.nbytes} bytes')
print(f'big_arr: {big_arr.nbytes} bytes')
print(f'huge_arr: {huge_arr.nbytes} bytes')
```

```python
# Test 4: String and dict
big_string = 'x' * 1000000
big_dict = {str(i): list(range(100)) for i in range(1000)}
print(f'big_string length: {len(big_string)}')
print(f'big_dict keys: {len(big_dict)}')
```

```python
# Test 5: Duplicate detection test
df_copy = df_big.copy()
df_slice = df_big[df_big.iloc[:, 0] > 0]
print('Created df_copy and df_slice from df_big')
```

```python
# Test 6: Delete a variable to free memory
# After running this, refresh the inspector to see huge_arr disappear
del huge_arr
import gc
gc.collect()
print('Deleted huge_arr')
```

## How to test

1. Open the **Vars** panel in the left sidebar (chart icon)
2. Run cells one by one with Shift+Enter
3. Click **refresh** in the inspector after each cell
4. Check that:
   - Variables appear with correct types
   - DataFrames show accurate sizes (not just shallow sys.getsizeof)
   - NumPy arrays show nbytes
   - Sizes are human-readable (KB, MB, GB)
   - Sorting by size puts biggest first
   - Deleting a variable with the X button removes it
5. Check the **status bar** at the bottom shows kernel RAM usage
6. After running cell 3 (huge_arr), RAM should jump noticeably
7. After running cell 6 (del huge_arr), RAM should drop
