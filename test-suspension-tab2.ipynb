```python
# Run this to allocate memory, then switch to Tab 1
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5000, 20), columns=[f'col_{i}' for i in range(20)])
result = df.describe()
print(f'Tab 2: DataFrame with {df.shape[0]} rows, {df.shape[1]} columns')
print(f'Memory: ~{df.memory_usage(deep=True).sum() / 1024 / 1024:.1f} MB')
result
```

```python
# After resuming from suspension, run this to verify state
# If Layer A (UI suspended): this should work, variables are preserved
# If Layer C (kernel stopped): this will fail because df is gone
try:
    print(f'df still exists! Shape: {df.shape}')
    print(f'result still exists! Columns: {list(result.columns)}')
except NameError as e:
    print(f'Variable lost: {e}')
    print('This is expected after Layer C (kernel was stopped and restarted)')
```
