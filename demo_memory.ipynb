# Saturn Memory Demo

This notebook demonstrates Saturn's memory visibility features.
Run each cell and watch the per-cell RAM deltas appear.

```python
import pandas as pd
import numpy as np
```

```python
# This cell creates a large DataFrame (~80 MB)
df = pd.DataFrame({
    'id': range(1_000_000),
    'value': np.random.randn(1_000_000),
    'category': np.random.choice(['A', 'B', 'C', 'D'], 1_000_000),
    'timestamp': pd.date_range('2024-01-01', periods=1_000_000, freq='s'),
})
print(f"DataFrame: {len(df):,} rows")
df.head()
```

```python
# This cell creates a large numpy array (~38 MB)
matrix = np.random.rand(5000, 1000)
print(f"Matrix shape: {matrix.shape}")
```

```python
# A small cell for contrast
result = df.groupby('category')['value'].mean()
print(result)
```

```python
# Clean up: delete the large array
del matrix
import gc
gc.collect()
print("Matrix deleted. Check the variable inspector - RAM should drop.")
```
